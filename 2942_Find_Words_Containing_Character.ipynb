```python
def findWordsContaining(words, x):
  #Initialise Output
  out = []
  #Initialise count, the number to add
  #to output list
  count = 0
  #Loop over words, add to output if x
  #value is in word
  for word in words:
    if x in word:
      out.append(count)
    else:
      pass
    count += 1
  return out
```

```python
words = ["leet","code"]
x = "e"
findWordsContaining(words, x)
```

```python
words = ["abc","bcd","aaaa","cbc"]
x = "a"
findWordsContaining(words, x)
```
