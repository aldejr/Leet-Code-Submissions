```python
def numberOfSteps(num):
  #Initialise Output
  counter = 0
  while num > 0:
    #Check for divisibility by 2
    if num % 2 == 0:
      num /= 2
    #If not possible, subtract 1
    else:
      num -= 1
    #Count the iteration
    counter += 1
  return counter
```

```python
num = 14
numberOfSteps(num)
```

```python
num = 8
numberOfSteps(num)
```

```python
num = 123
numberOfSteps(num)
```
