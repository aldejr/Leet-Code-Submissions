```python
def missingNumber(nums):
  #Found out this was unnecessary :)
  nums.sort()
  #Build proper list for checking
  x = len(nums) + 1
  complete_list = []
  for i in range(x):
    complete_list.append(i)
  #Initialise output
  out = 0
  #If we find an element of complete list
  #that is not in input nums, then we assign
  #it to the output variable
  for char in complete_list:
    if char not in nums:
      out = char
      break
    else:
      continue
  return out
```

```python
nums = [3,0,1]
missingNumber(nums)
```

```python
nums = [0,1]
missingNumber(nums)
```

```python
nums = [9,6,4,2,3,5,7,0,1]
missingNumber(nums)
```
