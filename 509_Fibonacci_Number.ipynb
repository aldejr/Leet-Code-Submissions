```python
def fib(n):
  #Hard Code to handle for 0 & 1
  if n == 0:
    add = 0
  elif n == 1:
    add = 1
  #add together the previous 2 numbers in sequence
  else:
    fib = [0, 1]
    for i in range(n-1):
      x = fib[-1] + fib[-2]
      fib.append(x)
    add = fib[-2] + fib[-3]
  return add
```
