```python
def maxFreqSum(s):
  #Establish vowels and build
  #string of consonants
  vowels = 'aeiou'
  alpha = string.ascii_lowercase
  consonants = ''
  for char in alpha:
    if char in vowels:
      pass
    else:
      consonants += char
  #Lists to store counts of vowels
  #and consonants
  v_list = []
  c_list = []
  check = list(set(s))
  for char in check:
    if char in consonants:
      c_list.append(s.count(char))
    else:
      v_list.append(s.count(char))
  #Identify max counts
  #handle for words with zero vowels or consonants
  try:
    v = max(v_list)
  except:
    if len(v_list) == 0:
      v = 0
  try:
    c = max(c_list)
  except:
    if len(c_list) == 0:
      c = 0
  #Produce sum
  return  c + v
```

```python
s = "successes"
maxFreqSum(s)
```
