```python
def alternatingSum(nums):
  #Intitialise Output
  out = 0
  #Iterate over nums, if index is even
  #add it to out, otherwise subtract from out
  for i in range(len(nums)):
    if i % 2 == 0:
      out += nums[i]
    else:
      out -= nums[i]
  return out
```

```python
nums = [1,3,5,7]
alternatingSum(nums)
```

```python
nums = [100]
alternatingSum(nums)
```
