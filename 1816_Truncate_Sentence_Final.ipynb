```python
def truncateSentence(s, k):
  #Create a list separated by spaces
  #truncated to the kth index
  slist = s.split(" ")[:k]
  #Return a new string
  return " ".join(slist)
```

```python
s = "Hello how are you Contestant"
k = 4
truncateSentence(s,k)
```

```python
s = "What is the solution to this problem"
k = 4
truncateSentence(s,k)
```

```python
s = "chopper is not a tanuki"
k = 5
truncateSentence(s, k)
```
