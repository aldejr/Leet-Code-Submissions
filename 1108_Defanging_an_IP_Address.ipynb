```python
def defangIPaddr(address):
  #Convert to list, this data type
  #is mutable
  address = list(address)
  #iterate over list, convert characters
  #if they meet the criteria
  for i in range(len(address)):
    if address[i] == ".":
      address[i] = '[.]'
    else:
      pass
  #convert back into string for output
  return ''.join(address)
```

```python
address = "1.1.1.1"
defangIPaddr(address)
```

```python
address = "255.100.50.0"
defangIPaddr(address)
```
