```python
def moveZeroes(nums):
  #Iterate through nums, if equal to zero
  #then remove/append to the end
  for num in nums:
    if num == 0:
      nums.remove(num)
      nums.append(num)
    else:
      pass
  return nums
```

```python
nums = [0,1,0,3,12]
moveZeroes(nums)
```

```python
nums = [0]
moveZeroes(nums)
```
