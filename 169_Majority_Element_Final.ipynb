```python
def majorityElement(nums):
  #Establish dictionary
  my_dict = {}
  #Unique elements to shorten
  #number of iterations for loop
  new_nums = set(nums)
  for nn in new_nums:
    #key/value pairs count/element
    x = nums.count(nn)
    my_dict[x] = nn
  #return max key
  return my_dict[max(my_dict.keys())]
```

```python
nums = [3,2,3]
majorityElement(nums)
```

```python
nums = [2,2,1,1,1,2,2]
majorityElement(nums)
```
