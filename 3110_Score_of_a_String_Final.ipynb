```python
def scoreOfString(s):
        #Initialise Output
        add = 0
        #Use range so that we can get the subtraction pairings
        for i in range(len(s)):
            try:
                #Perform subtraction, grab absolute value
                #of the output
                answer = abs(ord(s[i]) - ord(s[i+1]))
                add += answer
            except:
                break
        #Produce the sum of absolute value parings
        return add
```

```python
s = "hello"
scoreOfString(s)
```

```python
s = "zaz"
scoreOfString(s)
```
