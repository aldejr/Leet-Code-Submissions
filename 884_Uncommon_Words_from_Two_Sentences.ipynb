```python
def uncommonFromSentences(s1, s2):
  #Initialise Output
  out = []
  #Convert sentences to lists so as
  #to split them into individual words
  s1_list = s1.split(" ")
  s2_list = s2.split(" ")
  #Check words in S1 against S2
  for word in s1_list:
    if s1_list.count(word) == 1 and word not in s2_list:
      out.append(word)
    else:
      pass
  #Check words in S2 against S1
  for word in s2_list:
    if s2_list.count(word) == 1 and word not in s1_list:
      out.append(word)
    else:
      pass
  return out
```

```python
#Test Case 1
s1 = "this apple is sweet"
s2 = "this apple is sour"
uncommonFromSentences(s1, s2)
```

```python
#Test Case 2
s1 = "apple apple"
s2 = "banana"
uncommonFromSentences(s1, s2)
```
