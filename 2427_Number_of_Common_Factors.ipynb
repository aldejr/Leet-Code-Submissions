```python
def commonFactors(a, b):
  #Initialise Output
  count = 0
  #add 1 to count if both a and b
  #are divisible by i
  for i in range(1,a+1):
    if a % i == 0 and b % i == 0:
      count += 1
    else:
      pass
  return count
```

```python
#Test Case 1
a = 12
b = 6
commonFactors(a,b)
```

```python
#Test Case 2
a = 25
b = 30
commonFactors(a,b)
```
