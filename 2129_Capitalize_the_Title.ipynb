```python
def capitalizeTitle(title):
  #Convert input to list
  t_list = title.split(" ")
  #Initialise Output
  out = ''
  #For each word, check length and
  #make title/lower case as necessary
  for word in t_list:
    if len(word) == 1 or len(word) == 2:
      out += word.lower()
    else:
      out += word.title()
    out += " "
  #Make sure to remove trailing whitespaces
  return out.rstrip()
```

```python
#Test case 1
title = "capiTalIze tHe titLe"
capitalizeTitle(title)
```
