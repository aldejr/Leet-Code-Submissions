```python
def destCity(paths):
  #Initialise Output
  out = ''
  origins = []
  destinations = []
  #Build origin and destination lists
  for city in paths:
    beg = city[0]
    origins.append(beg)
    end = city[1]
    destinations.append(end)
  #Compare lists, extract the city that is
  #only in destinations
  for dest in destinations:
    if dest not in origins:
      out += dest
      break
    else:
      pass
  return out
```

```python
#Test Case 1
paths = [["London","New York"],["New York","Lima"],["Lima","Sao Paulo"]]
destCity(paths)
```

```python
#Test Case 2
paths = [["B","C"],["D","B"],["C","A"]]
destCity(paths)
```

```python
#Test Case 3
paths = [["A","Z"]]
destCity(paths)
```
