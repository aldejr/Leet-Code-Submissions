```python
def singleNumber(nums):
  #Initialise Output
  out = 0
  #Identify single count item
  #Break loop once found
  for num in nums:
    if nums.count(num) == 1:
      out = num
      break
    else:
      continue
  return out
```

```python
nums = [2,2,1]
singleNumber(nums)
```

```python
nums = [4,1,2,1,2]
singleNumber(nums)
```

```python
nums = [1]
singleNumber(nums)
```
