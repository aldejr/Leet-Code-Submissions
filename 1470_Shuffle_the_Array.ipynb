```python
def shuffle(nums, n):
  #Build X list, first n elements of nums
  x_list = []
  for i in range(n):
    x_list.append(nums[i])
  #Create Y list by removing x elements
  del nums[:n]
  #build output list out of combined/zipped x & Y
  #lists
  out = []
  check = zip(x_list, nums)
  for item in check:
    out.append(item[0])
    out.append(item[1])
  return out

```

```python
nums = [2,5,1,3,4,7]
n = 3
shuffle(nums, n)
```

```python
nums = [1,2,3,4,4,3,2,1]
n = 4
shuffle(nums, n)
```
