```python
def findGCD(nums):
  #Extract Max & Min
  nums_min = min(nums)
  nums_max = max(nums)
  #Obtain Divisors for minimum value
  min_divisors = []
  for i in range(1,nums_min +1):
    if nums_min % i == 0:
      min_divisors.append(i)
    else:
      pass
  #Obtain Divisors for max value
  max_divisors = []
  for i in range(1, nums_max+1):
    if nums_max % i == 0:
      max_divisors.append(i)
    else:
      pass
  #Build unified/common list of divisors
  out = []
  for div in min_divisors:
    if div in max_divisors:
      out.append(div)
    else:
      pass
  #Produce max from out list
  return max(out)
```

```python
nums = [2,5,6,9,10]
findGCD(nums)
```

```python
nums = [7,5,6,8,3]
findGCD(nums)
```

```python
nums = [3,3]
findGCD(nums)
```
