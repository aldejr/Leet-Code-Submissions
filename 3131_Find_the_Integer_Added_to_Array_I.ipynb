```python
def addedInteger(nums1, nums2):
  #Align the values by sorting the two
  #lists
  n1 = sorted(nums1)
  n2 = sorted(nums2)
  #Only need a single difference
  return n2[0] - n1[0]
```

```python
#Test Case 1
nums1 = [2,6,4]
nums2 = [9,7,5]
addedInteger(nums1,nums2)
```

```python
#Test Case 2
nums1 = [10]
nums2 = [5]
addedInteger(nums1,nums2)
```

```python
#Test Case 3
nums1 = [1,1,1,1]
nums2 = [1,1,1,1]
addedInteger(nums1,nums2)
```
