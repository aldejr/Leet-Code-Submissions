```python
def heightChecker(heights):
  #Create sorted list for comparison
  h_sort = sorted(heights)
  #Add 1 t out put only if the zipped tuple,
  #consisting of input and sorted check, don't
  #match
  counter = 0
  for i, j in zip(heights, h_sort):
    if i != j:
      counter += 1
    else:
      pass
  return counter
```

```python
#Test Case 1
heights = [1,2,3,4,5]
heightChecker(heights)
```

```python
#Test Case 2
heights = [5,1,2,3,4]
heightChecker(heights)
```

```python
#Test Case 3
heights = [1,1,4,2,1,3]
heightChecker(heights)
```
