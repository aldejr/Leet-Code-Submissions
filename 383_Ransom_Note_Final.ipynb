```python
def canConstruct(self, ransomNote, magazine):
        check = ''.join(list(set(ransomNote)))
        out = []
        for char in check:
            if magazine.count(char) >= ransomNote.count(char):
                out.append('Match')
            else:
                out.append('Miss')
        if list(set(out)) == ["Match"]:
            return True
        else:
            return False
```
