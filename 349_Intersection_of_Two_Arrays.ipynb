```python
def intersection(nums1, nums2):
  #Initialise Output
  out = []
  #Loop through nums1, add to output only
  #if it is also in nums2
  for num in nums1:
    if num in nums2:
      out.append(num)
    else:
      pass
  #Convert to set so that elements are unique
  return list(set(out))
```

```python
nums1 = [1,2,2,1]
nums2 = [2,2]
intersection(nums1, nums2)
```
