```python
def sumOfMultiples(n):
  #Initialise Output
  out = 0
  #Iterate over numbers in range
  #conditionally add to output if they meet the criteria
  for i in range(n+1):
    if i % 3 == 0:
      out += i
    elif i % 5 == 0:
      out += i
    elif i % 7 == 0:
      out += i
  #produce sum
  return out
```

```python
n = 7
sumOfMultiples(n)
```

```python
n = 10
sumOfMultiples(n)
```
