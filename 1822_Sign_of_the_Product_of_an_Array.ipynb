```python
def arraySign(nums):
  #Create a product, loop through
  #items in nums and multiply to x
  x = 1
  for num in nums:
    x *= num
  #Return result based on value of x
  if x < 0:
    return -1
  elif x > 0:
    return 1
  else:
    return 0
```

```python
#Test Case 1
nums = [-1,-2,-3,-4,3,2,1]
arraySign(nums)
```

```python
#Test Case 2
nums = [1,5,0,2,-3]
arraySign(nums)
```

```python
#Test Case 3
nums = [-1,1,-1,1,-1]
arraySign(nums)
```
