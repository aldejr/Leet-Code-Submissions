```python
def multiply(num1, num2):
  #Convert inputs to integers for operation
  #Convert product back into string for output
  return str(int(num1)*int(num2))
```

```python
#Test Case 1
num1 = "2"
num2 = "3"
multiply(num1, num2)
```
