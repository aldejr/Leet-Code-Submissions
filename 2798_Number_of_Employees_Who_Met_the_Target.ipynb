```python
def numberOfEmployeesWhoMetTarget(hours, target):
  #Initialise Output
  out = 0
  #If hours worked is greater than the target
  #add to output
  for hour in hours:
    if hour >= target:
      out += 1
    else:
      pass
  return out
```

```python
#Test Case 1
hours = [0,1,2,3,4]
target = 2
numberOfEmployeesWhoMetTarget(hours, target)
```

```python
#Test Case 2
hours = [5,1,4,2,2]
target = 6
numberOfEmployeesWhoMetTarget(hours, target)
```
