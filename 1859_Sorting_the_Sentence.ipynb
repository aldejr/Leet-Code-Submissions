```python
def sortSentence(s):
  #Convert to list
  s_list = s.split(" ")
  #build lists for words and their indices
  #we will eventually zip these together
  indices = []
  words = []
  for word in s_list:
    indices.append(int(word[-1]))
    words.append(word[:-1])
  zipped = sorted(zip(indices,words))
  #build output string based on sorted zipped list
  out = ''
  for item in zipped:
    out += item[1] +" "
  #Rstrip as there will be a trailing whitespace
  return out.rstrip()
```

```python
#Test Case 1
s = "is2 sentence4 This1 a3"
sortSentence(s)
```

```python
#Test Case 2
s = "Myself2 Me1 I4 and3"
sortSentence(s)
```
