```python
def missingMultiple(nums, k):
  x = 1
  while True:
    check = k * x
    if check not in nums:
      break
    else:
      x += 1
  return check
```

```python
#Test Case 1
nums = [8,2,3,4,6]
k = 2
missingMultiple(nums, k)
```

```python
#Test Case 2
nums = [1,4,7,10,15]
k = 5
missingMultiple(nums, k)
```
