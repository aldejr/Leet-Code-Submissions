```python
def findMissingElements(nums):
  #Establish list to check, this will have
  #no missing integers
  check = []
  for i in range(min(nums), max(nums)+1):
    check.append(i)
  #Create list to house the missing integers
  out = []
  for c in check:
    if c not in nums:
      out.append(c)
  else:
    pass
  #Produce list of missing integers
  return out
```

```python
nums = [1,4,2,5]
findMissingElements(nums)
```

```python
nums = [7,8,6,9]
findMissingElements(nums)
```

```python
nums = [5,1]
findMissingElements(nums)
```
