```python
def reversePrefix(word, ch):
  #Determine that ch is in word
  if ch in word:
    #Find ch index & reverse the prefix
    x = word.index(ch)
    prefix = word[:x+1][::-1]
    second_half = word[x+1:]
    #Create new word with reversed prefix
    return prefix + second_half
  #If ch is not in word, return the word
  else:
    return word
```

```python
#Test Case 1
word = "abcdefd"
ch = "d"
reversePrefix(word, ch)
```

```python
#Test Case 2
word = "xyxzxe"
ch = "z"
reversePrefix(word, ch)
```

```python
#Test Case 3
word = "abcd"
ch = "z"
reversePrefix(word, ch)
```
