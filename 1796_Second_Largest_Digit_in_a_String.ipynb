```python
def secondHighest(s):
  check = []
  for char in s:
    if char.isdigit() == True:
      check.append(int(char))
    else:
      pass
  check = list(set(check))
  check = sorted(check)
  try:
    return check[-2]
  except:
    return -1
```

```python
#Test Case 1
s = "dfa12321afd"
secondHighest(s)
```

```python
#Test Case 2
s = "abc1111"
secondHighest(s)
```

```python
#Test Case 3
s = "sjhtz8344"
secondHighest(s)
```
