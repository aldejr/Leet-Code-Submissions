```python
def findNumbers(nums):
  #Initialise Output
  counter = 0
  #Convert each element to string
  for i in range(len(nums)):
    x = len(str(nums[i]))
    #Check length of string element
    #Only add to counter if length is even
    if x % 2 == 0:
      counter += 1
    else:
      pass
  return counter
```

```python
#Test Case 1
nums = [12,345,2,6,7896]
findNumbers(nums)
```

```python
#Test Case 2
nums = [555,901,482,1771]
findNumbers(nums)
```
