```python
import string

def checkIfPangram(sentence):
  #create full alphabet string for test
  check1 = string.ascii_lowercase
  #convert sentence to sorted string of unique elements
  check2 = ''.join(sorted(set(sentence)))
  #If the string is equal to the alphabet, return True
  return check1 == check2
```

```python
#Test Case 1
sentence = "thequickbrownfoxjumpsoverthelazydog"
checkIfPangram(sentence)
```

```python
#Test Case 2
sentence = "leetcode"
checkIfPangram(sentence)
```
