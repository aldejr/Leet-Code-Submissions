```python
def firstPalindrome(words):
  #Initialize output
  out = ""
  #Check and see if each word is a palindrome
  #If it is, assign to output & break the cycle
  for word in words:
    if word == word[::-1]:
      out = word
      break
    else:
      pass
  #Produce palindromic or blank string
  return out
```

```python
words = ["abc","car","ada","racecar","cool"]
firstPalindrome(words)
```

```python
words = ["notapalindrome","racecar"]
firstPalindrome(words)
```

```python
words = ["def","ghi"]
firstPalindrome(words)
```
