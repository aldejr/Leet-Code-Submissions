```python
def searchInsert(nums, target):
  #First check, see if target is in the list
  if target in nums:
    #If so, return the index of target
    return nums.index(target)
  else:
    #If not, add target to the list
    #then provide the index of target
    nums.append(target)
    return sorted(nums).index(target)
```

```python
nums = [1,3,5,6]
target = 5
searchInsert(nums, target)
```
