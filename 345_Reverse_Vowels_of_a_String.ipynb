```python
def reverseVowels(s):
  #Establish vowels
  vowels = 'aeiou'
  #convert input to list so that it is mutable
  s_list = list(s)
  #Create list to hold vowels & their indices
  check = []
  indices = []
  #Build lists
  for i in range(len(s_list)):
    if s_list[i].lower() in vowels:
      indices.append(i)
      check.append(s_list[i])
    else:
      pass
  #Reverse indices
  indices.reverse()
  #Create Zip object and insert vowels at new, reversed
  #index positions
  conjoined = zip(indices,check)
  for tup in conjoined:
    s_list[tup[0]] = tup[1]
  return ''.join(s_list)
```

```python
#Test Case 1
s = "IceCreAm"
reverseVowels(s)
```
