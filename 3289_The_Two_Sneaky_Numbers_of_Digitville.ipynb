```python
def getSneakyNumbers(nums):
  n_set = set(nums)
  out = []
  for num in n_set:
    if nums.count(num) > 1:
      out.append(num)
    else:
      pass
  return out
```

```python
#Test Case 1
nums = [0,1,1,0]
getSneakyNumbers(nums)
```

```python
#Test Case 2
nums = [0,3,2,1,3,2]
getSneakyNumbers(nums)
```
