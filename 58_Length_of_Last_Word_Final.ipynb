```python
def lengthOfLastWord(s):
  #Create lists for processing
  #One for output, the other to handle input
  out = []
  new = s.split(' ')
  #Pluck only alphanumeric elements for output
  for item in new:
    if item.isalnum() == True:
      out.append(item)
  #produce length of last element
  return len(out[-1])
```

```python
s = "Hello World"
lengthOfLastWord(s)
```

```python
s = "   fly me   to   the moon  "
lengthOfLastWord(s)
```

```python
s = "luffy is still joyboy"
lengthOfLastWord(s)
```
