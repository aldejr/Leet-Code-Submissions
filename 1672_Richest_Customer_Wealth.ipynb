```python
def maximumWealth(accounts):
  #Initialise Output
  out = 0
  #Loop thru accounts, if the sum of
  #the account is greater than out, re-assign
  #out to that value
  for account in accounts:
    if sum(account) > out:
      out = sum(account)
    else:
      pass
  return out
```

```python
#Test Case 1
accounts = [[1,2,3],[3,2,1]]
maximumWealth(accounts)
```

```python
#Test Case 2
accounts = [[1,5],[7,3],[3,5]]
maximumWealth(accounts)
```

```python
#Test Case 3
accounts = [[2,8,7],[7,1,3],[1,9,5]]
maximumWealth(accounts)
```
