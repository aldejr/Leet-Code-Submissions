```python
def isBalanced(num):
  #Lists to Store evenly-indexed integers
  #and oddly-indexed integers
  evens = []
  odds = []
  #Iterate over nums, conditionally
  #append integers to the lists above
  #based on index positions
  for i in range(len(num)):
    if i % 2 == 0:
      evens.append(int(num[i]))
    else:
      odds.append(int(num[i]))
  #Perform boolean operation, make it the output
  return sum(evens) == sum(odds)
```

```python
num = "24123"
isBalanced(num)
```

```python
num = "1234"
isBalanced(num)
```
