```python
def findPeaks(mountain):
  #Initialise Output
  out = []
  #Apply conditions to mountain elements
  #If criteria are met, append to out
  for i in range(0,len(mountain)):
    try:
      if i == 0:
        pass
      elif mountain[i] > mountain[i-1] and mountain[i] > mountain[i+1]:
        out.append(i)
      else:
        pass
    except:
      pass
  #Produce out, either empty or with mountain indices
  return out
```
