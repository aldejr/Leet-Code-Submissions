```python
def toGoatLatin(sentence):
  s_list = sentence.split(' ')
  vowels = 'aeiou'
  out = []
  x = 1
  for word in s_list:
    if word[0].lower() in vowels:
      word += 'ma'
    else:
      word = word[1:] + word[0]
      word += 'ma'
    word += (x*"a")
    x += 1
    out.append(word)
  return ' '.join(out)
```

```python
#Test Case 1
sentence = "I speak Goat Latin"
toGoatLatin(sentence)
```

```python
#Test Case 2
sentence = "The quick brown fox jumped over the lazy dog"
toGoatLatin(sentence)
```
