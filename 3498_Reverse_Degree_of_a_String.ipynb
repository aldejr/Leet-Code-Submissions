```python

def reverseDegree(s):
  #Reverse the alphabet
  alpha = string.ascii_lowercase[::-1]
  #Initialise Output & Counter
  out = []
  count = 0
  #extract index from reversed alphabet
  #extract index from string for each character
  for char in s:
    first = alpha.index(char)+1
    second = count+1
    #Produce a product for output
    #add to out list
    product = first * second
    out.append(product)
    count += 1
  return sum(out)
```

```python
s = "abc"
reverseDegree(s)
```

```python
s = "zaza"
reverseDegree(s)
```
