```python
def rotateString(s, goal):
  #Initialise Output Default Value
  out = False
  #Convert s to list, use loop to
  #remove & append first character to list
  #break loop if the characters match goal
  s_list = list(s)
  for i in range(len(s_list)):
    r = s_list[0]
    s_list.remove(r)
    s_list.append(r)
    check = ''.join(s_list)
    if check == goal:
      out = True
      break
    else:
      continue
  return out
```

```python
#Test Case 1
s = "abcde"
goal = "cdeab"
rotateString(s,goal)
```

```python
#Test Case 2
s = "abcde"
goal = "abced"
rotateString(s,goal)
```
