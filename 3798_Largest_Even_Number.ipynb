```python
def largestEven(s):
  while True:
    try:
      if int(s) % 2 == 0:
        break
      else:
        s = s[:-1]
        continue
    except:
      s = ""
      break
  return s
```

```python
#Test Case 1
s = "1112"
largestEven(s)
```

```python
#Test Case 2
s = "221"
largestEven(s)
```

```python
#Test Case 3
s = "1"
largestEven(s)
```
