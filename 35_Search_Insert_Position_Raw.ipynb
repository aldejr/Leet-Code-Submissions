```python
nums = [1,3,5,6]
target = 2

if target in nums:
  print(nums.index(target))
else:
  print("Not there!")
```

```python
nums = [1,3,5,6]
target = 2

if target in nums:
  print(nums.index(target))
else:
  nums.append(target)
  print(sorted(nums).index(target))
```

```python
def searchInsert(nums, target):
  if target in nums:
    return nums.index(target)
  else:
    nums.append(target)
    return sorted(nums).index(target)
```

```python
nums = [1,3,5,6]
target = 7
searchInsert(nums, target)
```
