```python
def selfDividingNumbers(left, right):
  #Build string list of all numbers within range
  check = []
  for i in range(left, right +1):
    check.append(str(i))
  #Initialise Output
  out = []
  #add only self-divisible numbers to out
  for item in check:
    keep = True
    for char in item:
      if char == "0":
        keep = False
      elif int(item) % int(char) == 0:
        pass
      else:
        keep = False
    if keep == True:
      out.append(int(item))
    else:
      pass
  return out
```

```python
#Test Case 1
left = 1
right = 22
selfDividingNumbers(left, right)
```
