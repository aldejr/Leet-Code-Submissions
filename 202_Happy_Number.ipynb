```python
def isHappy(n):
  final = False
  for i in range(100):
    n_string = str(n)
    out = []
    for char in n_string:
      out.append(int(char))
    check = 0
    for o in out:
      check += o**2
    if check == 1:
      final = True
      break
    else:
      n = check
      continue
  return final
```

```python
#Test Case 1
n = 2
isHappy(n)
```

```python
#Test Case 2
n = 19
isHappy(n)
```
