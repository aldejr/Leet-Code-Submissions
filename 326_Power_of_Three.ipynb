```python
def isPowerOfThree(n):
  x = 0
  check = 3
  while True:
    new = check ** x
    x += 1
    if new >= n:
      break
    else:
      continue
  return new == n
```

```python
#Test Case 1
n = 27
isPowerOfThree(n)
```

```python
#Test Case 2
n = 0
isPowerOfThree(n)
```

```python
#Test Case 3
n = -1
isPowerOfThree(n)
```
