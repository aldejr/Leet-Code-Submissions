```python
def smallestEvenMultiple(n):
  #Initialise Output
  out = 1
  #Apply condition in while loop,
  #break if we find the first even multiple
  #continue if not, incrementally increasing the output
  #each time
  while True:
    if out % n == 0 and out % 2 == 0:
      break
    else:
      out += 1
      continue
  return out
```

```python
n = 5
smallestEvenMultiple(n)
```

```python
n = 6
smallestEvenMultiple(n)
```
