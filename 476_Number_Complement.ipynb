```python
def findComplement(num):
  #Convert input to binary
  check = bin(num)[2:]
  #Use for loop to invert the binary string
  out = ''
  for char in check:
    if char == '0':
      out += '1'
    else:
      out += '0'
  #Convert back to integer for output
  return int(out,2)
```

```python
#Test Case 1
num = 5
findComplement(num)
```

```python
#Test Case 2
num = 1
findComplement(num)
```
