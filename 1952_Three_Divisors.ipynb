```python
def isThree(n):
  #Initialise Output
  counter = 0
  #Add to counter only if it is a divisor
  for i in range(1,n+1):
    if n % i == 0:
      counter += 1
    else:
      pass
  #Produce true/false depending on counter value
  return counter == 3
```

```python
n = 2
isThree(2)
```

```python
n = 4
isThree(n)
```
