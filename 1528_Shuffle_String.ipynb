```python
def restoreString(s, indices):
  #Build a dummy list with matching length
  #as s & indices
  check = []
  for i in range(len(s)):
    check.append("-")
  #Join indices and characters, use loop
  #to insert characters into check at index
  #places in indices
  chars = zip(s,indices)
  for char in chars:
    check[char[1]] = char[0]
  return ''.join(check)
```

```python
#Test Case 1
s = "codeleet"
indices = [4,5,6,7,0,2,1,3]
restoreString(s, indices)
```

```python
#Test Case 2
s = "abc"
indices = [0,1,2]
restoreString(s, indices)
```
