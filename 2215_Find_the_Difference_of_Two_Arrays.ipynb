```python
def findDifference(nums1, nums2):
  #Initialise Output
  out = []
  #build first difference list
  check1 = []
  for num in nums1:
    if num not in nums2:
      check1.append(num)
    else:
      pass
  #Append unique list to output
  out.append(list(set(check1)))
  #build the second difference list
  check2 = []
  for num in nums2:
    if num not in nums1:
      check2.append(num)
    else:
      pass
  #append unique list to output
  out.append(list(set(check2)))
  return out
```

```python
nums1 = [1,2,3,3]
nums2 = [1,1,2,2]
findDifference(nums1, nums2)
```
