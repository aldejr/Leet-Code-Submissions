```python
def convertDateToBinary(date):
  #Convert to list, this makes the
  #input mutable
  check = date.split("-")
  #change each element to binary value
  for i in range(len(check)):
    check[i] = bin(int(check[i]))[2:]
  #Convert back into string for output
  return '-'.join(check)
```

```python
date = "2080-02-29"
convertDateToBinary(date)
```

```python
date = "1900-01-01"
convertDateToBinary(date)
```
