```python
def findTheDifference(s, t):
  #Initialise Output
  out = ''
  #Loop through t, pluck out the extra character
  #if it is not in s or if its count is higher in t
  #than it is in s
  for char in t:
    if char not in s:
      out = char
      break
    elif t.count(char) != s.count(char):
      out = char
      break
    else:
      continue
  return out
```

```python
s = "abcd"
t = "abcde"
findTheDifference(s,t)
```

```python
s = ""
t = "y"
findTheDifference(s,t)
```

```python
s = "a"
t = "aa"
findTheDifference(s,t)
```
