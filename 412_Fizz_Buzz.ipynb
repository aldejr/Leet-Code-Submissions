```python
def fizzBuzz(n):
  check = []
  for i in range(1, n+1):
    check.append(i)
  final = []
  for c in check:
    if c % 3 == 0 and c % 5 == 0:
      final.append("FizzBuzz")
    elif c % 3 == 0:
      final.append("Fizz")
    elif c % 5 == 0:
      final.append("Buzz")
    else:
      final.append(str(c))
  return final
```

```python
#Test Case 1
n = 3
fizzBuzz(n)
```

```python
#Test Case 2
n = 5
fizzBuzz(n)
```

```python
#Test Case 3
n = 15
fizzBuzz(n)
```
