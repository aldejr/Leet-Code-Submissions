```python
def numberGame(nums):
  out = []
  for i in range(int(len(nums)/2)):
    a_pick = min(nums)
    nums.remove(a_pick)
    b_pick = min(nums)
    nums.remove(b_pick)
    out.append(b_pick)
    out.append(a_pick)
  return out
```

```python
nums = [5,4,2,3]
numberGame(nums)
```

```python
nums = [2,5]
numberGame(nums)
```
