```python
def addDigits(num):
  #While loop too keep iterating
  #unless condition is met
  while True:
    #Convert to iterable that we can put
    #into a list
    n_string = str(num)
    n_list = []
    for char in n_string:
      n_list.append(int(char))
      num = sum(n_list)
    #If the sum of list elements is a 1 digit
    #number, then stop. otherwise continue
    if len(str(num)) == 1:
      break
    else:
      continue
  return num
```

```python
num = 38
addDigits(num)
```

```python
num = 0
addDigits(num)
```
