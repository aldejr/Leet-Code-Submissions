```python
def stringMatching(words):
  #Initialise Output
  out = []
  for i in range(len(words)):
    check = words[i]
    #Build list without check word
    reference = words[:i] + words[i+1:]
    #Add to output if it is a substring
    for ref in reference:
      if check in ref:
        out.append(check)
      else:
        pass
  #Produce unique list of substrings
  return list(set(out))
```

```python
#Test Case 1
words = ["leetcoder","leetcode","od","hamlet","am"]
stringMatching(words)
```

```python
#Test Case 2
words = ["blue","green","bu"]
stringMatching(words)
```
