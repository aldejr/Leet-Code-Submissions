```python
def thirdMax(nums):
  #Scenario for When list is less than
  #3 elements
  if len(nums) < 3:
    return max(nums)
  else:
    #Create distinct list of numbers
    check = set(nums)
    #Once again, handle for fewer than
    #3 distinct elements
    if len(check) < 3:
      return max(check)
    else:
      #Otherwise, return the 3rd highest number
      return list(sorted(check))[-3]
```

```python
nums = [3,2,1]
thirdMax(nums)
```

```python
nums = [1,2]
thirdMax(nums)
```

```python
nums = [2,2,3,1]
thirdMax(nums)
```

```python
nums = [1,1,2]
thirdMax(nums)
```

```python
nums = [-1,2,3]
thirdMax(nums)
```
