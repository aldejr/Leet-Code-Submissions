```python
def checkDivisibility(n):
  #Convert input to iterable
  n_string = str(n)
  #Initialize sum & build it with
  # n_string
  digit_sum = 0
  for char in n_string:
    digit_sum += int(char)
  #Initialise product & build it with
  #n_string
  digit_product = 1
  for char in n_string:
    digit_product *= int(char)
  #Produce output needed for boolean check
  check = digit_sum + digit_product
  return n % check == 0
```

```python
n = 99
checkDivisibility(n)
```

```python
n = 23
checkDivisibility(n)
```
