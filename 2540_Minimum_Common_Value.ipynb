```python
def getCommon(nums1, nums2):
  #Initialise Output
  out = []
  #Reduce length of inputs
  n1 = set(nums1)
  n2 = set(nums2)
  #Iterate over reduced-length set list
  #Add common elements to out
  for num in n2:
    if num in n1:
      out.append(num)
    else:
      pass
  #Produce minimum of out list or -1
  if len(out) == 0:
    return -1
  else:
    return min(out)
```

```python
nums1 = [1,2,3]
nums2 = [2,4]
getCommon(nums1, nums2)
```
