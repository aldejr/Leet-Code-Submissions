```python
def finalValueAfterOperations(operations):
  #Initialize Output
  out = 0
  #Loop through operations, conditionally add to
  #or subtract from out based on the list element
  for op in operations:
    if "+" in op:
      out += 1
    elif "-" in op:
      out -= 1
    else:
      pass
  return out
```

```python
operations = ["++X","++X","X++"]
finalValueAfterOperations(operations)
```

```python
operations = ["--X","X++","X++"]
finalValueAfterOperations(operations)
```

```python
operations = ["X++","++X","--X","X--"]
finalValueAfterOperations(operations)
```
