```python
def convertTemperature(celsius):
  #Initialize output
  out = []
  #Convert to kelvin & add to output
  kelvin = celsius + 273.15
  out.append(kelvin)
  #Convert to fahrenheit & add to output
  fahrenheit = celsius * 1.8 + 32
  out.append(fahrenheit)
  return out
```

```python
celsius = 36.50
convertTemperature(celsius)
```

```python
celsius = 122.11
convertTemperature(celsius)
```
